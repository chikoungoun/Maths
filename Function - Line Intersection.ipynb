```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1000)
f = np.arange(0, 1000)
g = np.sin(np.arange(0, 10, 0.01) * 2) * 1000

plt.plot(x, f, '-')
plt.plot(x, g, '-')
```

```python
idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
```

```python
idx
```

```python
plt.plot(x, f, '-')
plt.plot(x, g, '-')
plt.plot(x[idx], f[idx], 'ro')
```
